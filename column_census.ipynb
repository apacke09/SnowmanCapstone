### Get a census count of columns to help determine what I can cut now

```python
# needed for file operations
import os

# needed for data manipulation
import pandas as pd
```

```python
# get the list of weather station files that are within my lat/long box (output of previous step)
file_path = "F:/Libraries/My Documents/UM/Capstone/datafiles_in_range.txt"
ws_file_list = []
with open(file_path, 'r') as file:
    for line in file:
        ws_file_list.append(line.strip())
```

```python
directory = "F:/Libraries/My Documents/UM/Capstone/RawData/"
colcountdict = {}

for filename in ws_file_list:
    file_path = directory + filename
    df = pd.read_csv(file_path, low_memory=False)
    for col in df.columns:
        colcountdict[col] = colcountdict.get(col, 0) + 1
```

```python
print(colcountdict)
```

```python
sorted_dict_values = dict(sorted(colcountdict.items(), key=lambda item: item[1], reverse=True))
for key, value in sorted_dict_values.items():
    print(key, value)
```
