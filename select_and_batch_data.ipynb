### Read in all data that I'll be using, write to single csv file

```python
# needed for file operations
import os

# needed for data manipulation
import pandas as pd
```

```python
# get the list of weather station files that are meet minimum requirements (output of previous step)
file_path = "F:/Libraries/My Documents/UM/Capstone/datafiles_in_range_meet_data_req.txt"
ws_file_list = []
with open(file_path, 'r') as file:
    for line in file:
        ws_file_list.append(line.strip())
```

```python
# for each of these weather station files, concat onto cumulative dataframe
dir_with_raw_data_files = "F:/Libraries/My Documents/UM/Capstone/RawData"
columns_to_keep = ['STATION', 'DATE', 'LATITUDE', 'LONGITUDE', 'ELEVATION', \
                   'PRCP', 'SNOW', 'SNWD', 'TMAX', 'TMIN']
                   
batch_number = 0
batch_ws_df = pd.DataFrame()
for i in range(0, len(ws_file_list)):
    file = ws_file_list[i]
    file_path = os.path.join(dir_with_raw_data_files, file)
    df = pd.read_csv(file_path, low_memory=False)
    columns_to_keep_that_exist = [col for col in columns_to_keep if col in df.columns]
    df = df[columns_to_keep_that_exist]
    batch_ws_df = pd.concat([batch_ws_df, df])
    if i % 100 == 0:
        print(f"Processing file {i} of {len(ws_file_list)}")
    if (i+1) % 1000 == 0 or i == len(ws_file_list)-1:
        batch_filename = "F:/Libraries/My Documents/UM/Capstone/ws_batch_" + str(batch_number) + ".csv"
        batch_ws_df.to_csv(batch_filename, index=False)
        batch_number += 1
        batch_ws_df = pd.DataFrame()


```
